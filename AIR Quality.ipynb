```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['datum'] = label_encoder.fit_transform(dataset['datum'])
dataset['sample_duration'] = label_encoder.fit_transform(dataset['sample_duration'])
```

```python
datum = dataset['datum']
sample_duration = dataset['sample_duration']
```

```python
import matplotlib.pyplot as plt
yearvsfiftypercentile = dataset.groupby('year')['fifty_percentile'].mean()
plt.figure(figsize=(10, 6))
plt.plot(yearvsfiftypercentile.index, yearvsfiftypercentile.values, marker='*')
plt.xlabel('Year')
plt.ylabel('fifty_percentile')
plt.title('year vs fifty_percentile')
plt.show()
```

```python
yearvsseventyfivepercentile = dataset.groupby('year')['seventy_five_percentile'].mean()
plt.figure(figsize=(10, 6))
plt.plot(yearvsseventyfivepercentile.index, yearvsseventyfivepercentile.values, marker='*')
plt.xlabel('Year')
plt.ylabel('seventy five percentile')
plt.title('year vs seventy five percentile')
plt.show()
```

```python
# Plotting a line chart
plt.plot(dataset['state_name'], label='state_name')
plt.plot(dataset['fifty_percentile'], label='fifty_percentile')
plt.xlabel('State Name')
plt.ylabel('Fifty percentile')
plt.legend()
plt.title('Comparison between State Name and Fifty Percentile')
plt.show()
```

```python

# Plotting a scatter plot
plt.scatter(dataset['year'], dataset['seventy_five_percentile'])
plt.xlabel('year')
plt.ylabel('seventy_file_percentile')
plt.title('Comparison between year and seventy file percentile')
plt.show()
```

```python
selected_features=dataset[['datum','sample_duration','year','parameter_code']]
label = dataset[['fifty_percentile']]
```

```python
corr = selected_features.corr()
print(corr)
```

```python
import seaborn as sns
sns.heatmap(corr)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, label, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor()
clf.fit(X_train, y_train)
```

```python
x_test = scaler.transform(X_test)
```

```python
predict = clf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(y_test,predict)
print(error)
```
